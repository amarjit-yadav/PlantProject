```python
path='/home/drowsy/Downloads/data/train/images'

def get_names(path):
    folder = path

    sub_folders = [name for name in os.listdir(folder) if os.path.isdir(os.path.join(folder, name))]

#     print(sub_folders)
    
    return pd.DataFrame(sub_folders,columns=['classid'])

def map_meta(full_data,curr_data):
    merged_data=full_data.merge(curr_data,on='classid',how ='inner')
    return merged_data
```

```python
curr_df=get_names(path)
curr_df
```
